```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = plt.imread('data/image.jpg')
image.shape
```

```python
i,j = 1000,159
image[i,j]
```

```python
A = np.mean(image, -1) / 256
A.shape
```

```python
A[i,j]
```

```python
plt.imshow(A)
plt.show()
```

```python
A_hat = np.fft.fft2(A)
A_hat.shape, A_hat.dtype
```

```python
x, y = np.meshgrid(np.arange(0, A_hat.shape[1],1), np.arange(0, A_hat.shape[0],1))
x.shape, y.shape
```

```python
plt.contour(x, y, np.abs(A_hat))
```
